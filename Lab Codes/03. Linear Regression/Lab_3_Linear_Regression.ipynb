Assume you have:

X = -1, 0, 4, 3, 2, 1

Y = -3, -1, 7, 5, 3, 1

What's the relation between X and Y?

It's Y = 2X - 1.


How do we train a neuron to learn this relation?

We will explore linear regression in this code.

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```
#Defining a layer of connected layers
model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
```

```
#Defining the optimizer and loss function
model.compile(optimizer='sgd', loss = 'mean_squared_error')
```

```
xs = np.array([-1.0, 0.0, 4.0, 3.0, 2.0, 1.0], dtype = float)
ys = np.array([-3.0, -1.0, 7.0, 5.0, 3.0, 1.0],dtype = float)
```

```
#Model fitting
model.fit(xs, ys, epochs = 500)
```

```
#For X = 10.0, Y should be 19.0
print(model.predict([10.0]))
```
